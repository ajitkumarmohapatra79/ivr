```python
def get_dim(embedding):
    embed_string = str(embedding)
#     print('embed_string={}'.format(embed_string))
    comma = embed_string.index(',')
    dim = embed_string[comma+1:-1].strip()
#     print('dim={}'.format(dim))
    return int(dim)
```

```python
get_dim(cat_lookup_df.iloc[4]['embed'])
```

```python
total_dim = 0

for embed in cat_lookup_df['embed'].values:
    total_dim += get_dim(embed)
```
